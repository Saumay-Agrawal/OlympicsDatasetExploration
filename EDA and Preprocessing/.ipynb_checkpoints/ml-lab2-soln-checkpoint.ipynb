```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb
import math
```

```python
data = pd.read_csv('olympic-data.csv')
data.head(5)
```

```python
data.info()
```

```python
regions = pd.read_csv('region-data.csv')
regions.head(5)
```

```python
del regions['notes']
regions.rename(columns = {'region': 'Region'}, inplace=True)
regions.head(5)
```

```python
df = pd.merge(data, regions, on='NOC')
df.head(5)
```

```python
df.info()
```

```python
df.loc[df['Region'].isnull()].head(21)
```

```python
df.loc[df['Team']=='Refugee Olympic Athletes', 'Region'] = 'Refugee Olympic Athletes'
df.loc[(df['Region'].isnull()) & (df['Team']=='Tuvalu'), 'Region'] = 'Tuvalu'
df = df.loc[(df['Region'].notnull()) & (df.NOC!='UNK')]
df.loc[(df['Region']=='Refugee Olympic Athletes') | (df['Region']=='Tuvalu')]
```

```python
df.loc[df['Medal'].isnull(), 'Medal'] = 'None'
df.loc[df['Medal'] == 'None']
```

```python
for index, row in df.iterrows():
    game = '{} {}'.format(row['Year'], row['Season'])
    if game!=row['Games']:
        print(row)
```

```python
del df['Games']
df.head(5)
```

```python
for region in df['Region'].value_counts().index:
    attributes = ['Age', 'Height', 'Weight']
    for attr in attributes:
        male_mean = df.loc[(df.Region==region) & (df.Sex=='M') & (df[attr].notnull()), attr].mean()
        female_mean = df.loc[(df.Region==region) & (df.Sex=='F') & (df[attr].notnull()), attr].mean()
        if not math.isnan(male_mean):
            df.loc[(df.Region==region) & (df.Sex=='M') & (df[attr].isnull()), attr] = round(male_mean)
        if not math.isnan(female_mean):
            df.loc[(df.Region==region) & (df.Sex=='F') & (df[attr].isnull()), attr] = round(female_mean)
```

```python
df.loc[df.Height.isnull()]
```

```python
df.loc[(df.Region=='Tuvalu') & (df.Sex=='F') & (df.Height.isnull()), 'Height'] = df.loc[df['Sex']=='F', 'Height'].mean()
df.info()
```

```python
df.to_csv('prepd-data.csv')
```

```python
df.describe()
```

```python
sb.countplot(data=df, x='Year')
```

```python
years = df.Year.unique()
years = np.sort(years)
print(years)
```

```python
df20 = df.loc[df.Year.isin(years[-5:])]
sb.countplot(data=df20, x='Year')
df20.to_csv('prepd-data-20y.csv')
```

```python
df40 = df.loc[df.Year.isin(years[-11:])]
sb.countplot(data=df40, x='Year')
df40.to_csv('prepd-data-40y.csv')
```

```python
mat = []
for region in df20['Region'].unique():
    count = df.loc[(df.Medal!='None') | (df.Region==region), 'Medal'].count()
    mat.append([region, count])
df_medal = pd.DataFrame(mat, columns=['Region', 'Medal'])
df_medal.head(10)
```

```python
df_medal.nlargest(10, 'Medal')
```

```python
top_regions = df_medal.nlargest(10, 'Medal')['Region'].unique()
print(top_regions)
```

```python
mat = []
for sport in df20['Sport'].unique():
    count = df.loc[(df.Sport==sport)].shape[0]
    mat.append([sport, count])
df_sport = pd.DataFrame(mat, columns=['Sport', 'Athletes'])
top_sports = df_sport.nlargest(10, 'Athletes')['Sport'].unique()
print(top_sports)
```

```python
sb.relplot(data=df20.loc[(df.Medal!='None') & (df.Sport.isin(top_sports))], x='Height', y='Weight', row='Sport', col='Medal', hue='Sex', height=2, aspect=2)
```

```python
sb.catplot(kind='violin', data=df20.loc[(df.Medal!='None') & (df.Sport.isin(top_sports))], x='Medal', y='Age', row='Sport', col='Year', hue='Sex', split=True, height=2, aspect=2)
```

```python
sb.catplot(kind='count', data=df20.loc[(df.Medal!='None') & (df.Region.isin(top_regions))], x='Medal', hue='Sex', row='Region', col='Year', height=2, aspect=2)
```
